```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#1.
years = [1951, 1961, 1971, 1981, 1991, 2001, 2011]
population = [
    361088000,
    439235000,
    548160000,
    683329000,
    846387888,
    1028737436,
    1210726932
]
plt.figure(figsize=(8,5))
plt.plot(years, population, marker='o', linestyle='-', color='b', label="Population Growth")
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Growth in India (1951 - 2011)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#2.


x = np.linspace(0, 2 * np.pi, 100)   
y = np.sin(x)                        
plt.figure(figsize=(8,5))
plt.plot(x, y, color="red", label="sin(x)")
plt.xlabel("x values (radians)")
plt.ylabel("sin(x)")
plt.title("Sine Curve using NumPy")
plt.axhline(0, color="black", linewidth=0.5)  
plt.axvline(0, color="black", linewidth=0.5)  
plt.grid(True)
plt.legend()
plt.show()
```

```python
#3

x = np.linspace(0, 2 * np.pi, 100) 
y = np.cos(x)                        
plt.figure(figsize=(8,5))
plt.plot(x, y, color="red", label="cos(x)")
plt.xlabel("x values (radians)")
plt.ylabel("cos(x)")
plt.title("Cosine Curve using NumPy")
plt.axhline(0, color="black", linewidth=0.5)  # x-axis
plt.axvline(0, color="black", linewidth=0.5)  # y-axis
plt.grid(True)
plt.legend()
plt.show()
```

```python
#4
classes = ["6", "7", "8", "9", "10", "11", "12"]
boys = [30, 35, 28, 32, 34, 29, 31] 

plt.figure(figsize=(8,5))
plt.bar(classes, boys, color="orange", edgecolor="black")
plt.xlabel("Class")
plt.ylabel("Number of Boys")
plt.title("Number of Boys in Classes 6 to 12")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#5
subjects = ["Maths", "Science", "English", "Social", "Hindi", "Computer"]
marks = [88, 92, 76, 85, 80, 95]  
plt.figure(figsize=(8,5))
plt.bar(subjects, marks, color="purple", edgecolor="black")
plt.xlabel("Subjects")
plt.ylabel("Marks Scored")
plt.title("Marks Scored in Different Subjects")
plt.ylim(0, 100) 
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#6
arrival_times = [7.1, 7.4, 7.5, 7.8, 8.0, 8.3, 8.5, 8.7, 
                 9.0, 9.2, 9.5, 9.7, 10.1, 10.4, 10.8, 
                 11.0, 11.3, 11.5, 11.7, 11.9]
plt.figure(figsize=(8,5))
plt.hist(arrival_times, bins=[7,8,9,10,11,12], color="lightblue", edgecolor="black")
plt.xlabel("Time of Arrival (Hours)")
plt.ylabel("Number of Employees")
plt.title("Employee Arrivals between 7 AM - 12 Noon (1-hour intervals)")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
